```python
import pandas as pd
from pandas import DataFrame
from pandas import Series
import numpy as np
import matplotlib.pyplot as plt


names1880 = pd.read_csv('/home/ubuntu/pydata/ch02/names/yob1880.txt', names = ['name', 'sex', 'births'], header = None)

names1880.head()

```

```python
names1880.groupby('sex').births.sum()

years = range(1880, 2011)

pieces = []

columns = ['name', 'sex', 'births']

for year in years:
    path = '/home/ubuntu/pydata/ch02/names/yob%d.txt' % year
    frame = pd.read_csv(path, names = columns)
    frame['year'] = year
    pieces.append(frame)
    
pieces    
names = pd.concat(pieces, ignore_index=True)
?pd.concat

names
```

```python
total_births = names.pivot_table('births', rows= 'year', 
                                 cols = 'sex', aggfunc = sum)
print total_births.tail()

total_births.plot(title = 'Total births by sex and year')
```

```python
names

test = names.groupby(['year', 'sex'])

def add_prop(group):
    # Integer division floors
    births = group.births.astype(float)
    
    group['prop'] = births / births.sum()
    return group

names = names.groupby(['year', 'sex']).apply(add_prop)
```

```python
def get_top1000(group):
    return group.sort_index(by='births', ascending = False)[:1000]
grouped = names.groupby(['year','sex'])
top1000 = grouped.apply(get_top1000)



```

```python
boys = top1000[top1000.sex == 'M']
girls = top1000.ix[top1000.sex == 'F',:]

boys
girls
```

```python
total_births = top1000.pivot_table('births', rows='year', cols= 'name',
                                  aggfunc = 'sum')


total_births
```

```python
subset = total_births[['John', 'Harry', 'Mary', 'Marilyn']]

subset.plot(subplots=True, figsize=(12, 10), grid=True, 
            title='Number of births per year')
```

```python
total_births.sum(axis =1).plot(title='Total birth over years')

table = top1000.pivot_table('prop', rows='year', 
                            cols='sex', aggfunc=sum)
table.plot(title='Sum of table1000.prop by year and sex', figsize=(10,4),
           yticks=np.linspace(0, 1.2, 13), xticks=range(1880, 2020, 10))
```

```python
def get_quantile_count(group, q=0.5):
    group = group.sort_index(by='prop', ascending=False)
    return group.prop.cumsum().searchsorted(0.5) + 1

diversity = top1000.groupby(['year', 'sex']).apply(get_quantile_count)
diversity = diversity.unstack('sex')
diversity.head()

diversity.plot(title='Number of popular names in top 50%')
```
